Curves to graph

- $y = 0n+1$ and label the curve "O(1)"
- $y = log(n)$ and label the curve "O(log n)"
- $y = n$ and label the curve "O(n)"
- $y = n * log(n)$ and label it "O(n log n)"
- $y = n^2$ and label it "O(n^2)"
- $y = 2^n$ and label it "O(2^n)"
- $y = n!$ and label it "O(n!)"
- $y = n^n$ and label it "O(n^n)"

```python
import numpy as np
import matplotlib.pyplot as plt
from math import log, factorial
```

```python
x = np.random.randint(1, 100, 100)
x = np.sort(x)
x
```

$y=0n+1$ and label the curve "O(1)"

```python
y = [n * 0 + 1 for n in x]
```

```python
plt.plot(x,y)
plt.title('$O(1)$')
```

$y1 = log(n)$ and label the curve "O(log n)"


```python
y1 = [log(float(n)) for n in x]
```

```python
plt.plot(x,y1)
plt.title('$O(log n)$')
```

$y2 = n$ and label the curve "O(n)"

```python
y2 = [n * 2 for n in x]
```

```python
plt.plot(x,y2)
plt.title('$O(n)$')
```

$y3 = n * log(n)$ and label it "O(n log n)"

```python
y3 = [n * log(n) for n in x]
```

```python
plt.plot(x,y3)
plt.title('$O(n log n)$')
```

$y4 = n^2$ and label it "O(n^2)"

```python
y4 = [n**2 for n in x]
```

```python
plt.plot(x,y4)
plt.title('$O(n^2)$')
```

$y5 = 2^n$ and label it "O(2^n)"

```python
y5 = [2 ** n for n in x]
```

```python
plt.plot(x,y5)
plt.title('$O(2^n)$')
#This appears to go out of range
```

$y6 = n!$ and label it "O(n!)"

```python
y6 = [factorial(n) for n in x]
```

```python
plt.plot(x,y6)
plt.title('$O(n!)$')
```

$y7 = n^n$ and label it "O(n^n)"

```python
y7 = [n**n for n in x]
```

```python
plt.scatter(x,y7)
plt.title('$O(n^n)$')
```
